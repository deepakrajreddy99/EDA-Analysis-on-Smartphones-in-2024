```python
import matplotlib.pyplot as plt

# Histograms for numerical columns
data.hist(bins=20, figsize=(15, 10))
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
# Scatter plots for Price vs. selected features
features = ['ram', 'storage', 'BATTERY']  # Replace with relevant features in your dataset
for feature in features:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=data[feature], y=data['price'])  # Replace 'Price' with your column name
    plt.title(f"{feature} vs price")
    plt.xlabel(feature)
    plt.ylabel("price")
    plt.show()
```

```python
# Count plot for two categorical columns, e.g., Brand vs Operating System
sns.countplot(x=data['phone_brand'], hue=data['price'])  # Replace 'Brand' and 'Operating_System' with actual column names
plt.title("Distribution of price within Brand")
plt.xticks(rotation=60)
plt.show()


```

```python
# Boxplot for Storage vs Brand
sns.boxplot(x=data['phone_brand'], y=data['ram'])  # Replace 'Brand' and 'Storage' with actual column names
plt.title("ram Distribution by Brand")
plt.xticks(rotation=45)
plt.show()
```
